```python
import pandas as pd
import numpy as np
```

# Importação e tratamento de dados COVID

```python
df_cv = pd.read_csv('base_covid.csv')
```

```python
df_cv.columns
```

```python
df_cv['periodo']=df_cv['periodo'].astype(str)
```

```python
df_cv.loc[df_cv['NOME MUNICÍPIO']=='ACRELANDIA']
```

```python
df_cvnovo = df_cv.groupby(['NOME MUNICÍPIO','estado','periodo','populacaoTCU2019'],as_index=False).sum(['casosNovos','obitosNovos'])
```

```python
df_cvnovo.head()
```

```python
df_cvnovo['year']=df_cvnovo['periodo'].apply(lambda x: str(x)[:4])
df_cvnovo['concatenar']=df_cvnovo['periodo']+df_cvnovo['NOME MUNICÍPIO']+df_cvnovo['estado']
```

```python
df_cvnovo = df_cvnovo.loc[df_cvnovo['year']=='2020']
```

```python
df_cvnovo.columns
```

```python
df_cvnovo['populacaoTCU2019']=df_cvnovo['populacaoTCU2019'].astype(int)
```

```python
df_cvnovo = df_cvnovo[['NOME MUNICÍPIO', 'estado', 'periodo', 'populacaoTCU2019','CÓDIGO MUNICÍPIO IBGE', 'casosNovos', 'obitosNovos',
       'year', 'concatenar']]
```

# Tratamento de dados Base Auxilio

```python
df_auxilio = pd.read_csv('base_auxilio.csv')
df_auxilio.head()
```

```python
df_novo = df_auxilio.groupby(['NOME MUNICÍPIO','UF','MÊS DISPONIBILIZAÇÃO'],as_index=False).sum()
```

```python
df_novo.head()
```

```python
df_novo.loc[df_novo['NOME MUNICÍPIO']=='BELEM']
```

```python
df_novo.columns
```

```python
df_novo['concatenar']=df_novo['MÊS DISPONIBILIZAÇÃO']+df_novo['NOME MUNICÍPIO']+df_novo['UF']
```

```python
df_novo= df_novo[['NOME MUNICÍPIO', 'UF', 'MÊS DISPONIBILIZAÇÃO','CÓDIGO MUNICÍPIO IBGE', 'soma_valor_beneficio',
       'count_valor_beneficio','concatenar']]
```

```python
df_novo.head()
```

```python
df_novo['soma_valor_beneficio']=df_novo['soma_valor_beneficio'].astype(int)
```

```python
df_novo.to_csv('Base_aux_final.csv',sep=";")
df_cvnovo.to_csv('Base_cvfinal.csv',sep=";")
```

# Mesclando Data Frame COVID e de Auxilio Emergencial

```python
df = pd.merge (df_novo,df_cvnovo, left_on=['MÊS DISPONIBILIZAÇÃO','NOME MUNICÍPIO'], right_on=['periodo','NOME MUNICÍPIO'],how='left')
```

```python
df.head()
```

```python
df.columns
```

```python
df= df[['NOME MUNICÍPIO', 'UF','MÊS DISPONIBILIZAÇÃO', 'CÓDIGO MUNICÍPIO IBGE_x', 'soma_valor_beneficio',
       'count_valor_beneficio', 'periodo', 'populacaoTCU2019','CÓDIGO MUNICÍPIO IBGE_y', 'casosNovos',
       'obitosNovos']]
```

```python
df.head()
```

```python
df['ben_populacao'] = df['count_valor_beneficio']/df['populacaoTCU2019'] 
df['percapita'] = df['soma_valor_beneficio']/df['populacaoTCU2019'] 
```

```python
df['soma_valor_beneficio']=df['soma_valor_beneficio'].astype(int)
```

```python
df['year']=df['periodo'].apply(lambda x: str(x)[:4])
```

```python
df['mes']=df['periodo'].apply(lambda x: str(x)[-2:])
```

```python
df['mes']=df['mes'].astype(str)
```

```python
df.head()
```

```python
df['percapita']=df['percapita'].astype(str)
```

```python
df['percapita'] = df['percapita'].apply(lambda x: x.replace('.',',') )
```

```python
df['ben_populacao']=df['ben_populacao'].astype(str)
```

```python
df['ben_populacao'] = df['ben_populacao'].apply(lambda x: x.replace('.',',') )
```

```python
df['casosNovos']=df['casosNovos'].astype(str)
```

```python
df['casosNovos']=df['casosNovos'].apply(lambda x: x.replace('.',',') )
```

```python
df['populacaoTCU2019']=df['populacaoTCU2019'].astype(str)
```

```python
df['populacaoTCU2019']=df['populacaoTCU2019'].apply(lambda x: x.replace('.',',') )
```

```python
df['obitosNovos']=df['obitosNovos'].astype(str)
```

```python
df['obitosNovos']=df['obitosNovos'].apply(lambda x: x.replace('.',',') )
```
